### Solution

```python
from sklearn import svm, datasets
from sklearn.svm import SVC
from sklearn.datasets import load_iris
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics.cluster import completeness_score
```

```python
#Prepare data
iris = datasets.load_iris()
X, y = iris.data, iris.target
XsampleTest = np.array([[3,5,4,2],])
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.35,random_state=0)

#Build model
model = svm.SVC(kernel="poly", degree=3, gamma="auto")
model.fit(X_train,y_train)
result = model.predict(XsampleTest)
print(iris.target_names[result])
```

```python
#Model validation

#Run model
result = model.predict(X_test)


#Generate confusion matrix
print("Confusion matrix: \n" ,confusion_matrix(y_test, result))

#Generate purity score
contingency_matrix = metrics.cluster.contingency_matrix(y_test, result)
print("Purity: ", round(np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix)*100,2),"%")

#Generate completness score
print("Completness: ", round(completeness_score(y_test, result)*100,2),"%")

#Show raw data success
predicted = iris.target_names[result]
actual = iris.target_names[y_test]
df = pd.DataFrame(list(zip(predicted, actual)),columns =['Predicted', 'Actual'])
df['Result'] = np.where(df['Predicted']== df['Actual'], "Correct", "Incorrect")
def coloring(series):    
    highlight = 'background-color: #6a7374;'
    default = ''
    return [highlight if row == "Incorrect" else default for row in series]
    
df.style.apply(coloring, subset=['Result'])
```
